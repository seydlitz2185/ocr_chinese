# Formal 目录 PDF OCR

使用 PaddleOCR 对 `../formal/` 目录中的一个 PDF 执行 OCR，并汇总识别结果。

```python
# 如需安装依赖，请取消注释后运行
# %pip install -q paddlepaddle paddleocr pandas tqdm
```

```python
from pathlib import Path
import pandas as pd
from paddleocr import PaddleOCR

ocr = PaddleOCR(use_textline_orientation=True, lang="ch")
```

```python
PDF_DIR = Path("../formal")
pdf_files = sorted(PDF_DIR.glob("*.pdf"))

if not pdf_files:
    raise FileNotFoundError(f"未在 {PDF_DIR} 找到 PDF 文件")

pdf_path = pdf_files[0]  # 如需指定文件可改为: Path('../formal/你的文件名.pdf')
print(f"将处理 PDF: {pdf_path}")
```

```python
from tqdm.auto import tqdm

results = ocr.predict(str(pdf_path))
total_pages = len(results)
print(f"总页数: {total_pages}")

rows = []
for page_idx, page in enumerate(
    tqdm(results, total=total_pages, desc="OCR 页处理进度", unit="页"),
    start=1,
 ):
    texts = page.get("rec_texts") or []
    scores = page.get("rec_scores") or []
    boxes = page.get("dt_polys") or []

    print(f"第 {page_idx} 页识别到 {len(texts)} 个文本区域")

    for text, score, box in zip(texts, scores, boxes):
        rows.append({
            "页码": page_idx,
            "识别文字": text,
            "置信度": round(float(score), 4),
            "左上角坐标": box[0],
            "右下角坐标": box[2],
        })

df = pd.DataFrame(rows)
df.head(20)
```

```python
output_csv = Path("./formal_pdf_ocr_result.csv")
output_txt = Path("./formal_pdf_ocr_text.txt")

df.to_csv(output_csv, index=False, encoding="utf-8-sig")
output_txt.write_text("\n".join(df["识别文字"].astype(str).tolist()), encoding="utf-8")

print(f"CSV 已保存: {output_csv.resolve()}")
print(f"文本已保存: {output_txt.resolve()}")
print(f"总识别条数: {len(df)}")
```
